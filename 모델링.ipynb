```python
from konlpy.tag import Okt

def get_lemmas(text_list):
    okt = Okt()
    result = []
    for text in text_list:
        lemmas = okt.pos(text, norm=True, stem=True)
        lemmas = [lemma[0] if (lemma[1] not in ['Josa', 'PreEomi', 'Eomi', 'Suffix', 'Punctuation']) else '' for lemma in lemmas]
        result.append(' '.join(lemmas))
    return result

# 입력 텍스트 리스트
text_list = df['review'][0]

# 표제어 추출
lemmatized_texts = get_lemmas(text_list)
for text in lemmatized_texts:
    print(text)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# 모델을 정의합니다
model = Sequential()
# 임베딩 레이어를 추가합니다. 이 레이어는 어휘 사전의 크기를 5000으로, 임베딩 벡터의 차원을 64로, 입력 시퀀스의 길이를 100으로 설정합니다.
model.add(Embedding(input_dim=5000, output_dim=64, input_length=100))
# LSTM 레이어를 추가합니다. LSTM 유닛의 수는 64로 설정합니다.
model.add(LSTM(64))
# 완전 연결 레이어를 추가합니다. 출력 노드의 수는 레이블의 수와 동일하게 설정하고, 활성화 함수로는 시그모이드를 사용합니다.
model.add(Dense(labels.shape[1], activation='sigmoid'))

# 모델을 컴파일합니다. 손실 함수로는 이진 크로스 엔트로피를, 최적화 알고리즘으로는 아담을, 평가 지표로는 정확도를 사용합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 데이터를 훈련 세트와 테스트 세트로 분할합니다. 테스트 세트의 비율은 20%로 설정합니다.
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)

# 모델을 훈련합니다. 에포크 수는 10으로 설정하고, 검증 데이터로 테스트 세트를 사용합니다.
model.fit(X_train, y_train, epochs=500, validation_data=(X_test, y_test))
```
