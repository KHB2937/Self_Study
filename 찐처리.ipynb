```python
import re

def text_cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
    result = hangul.sub('', text)
    return result

df['review'] = df['review'].apply(lambda x: text_cleaning(' '.join(x)) if isinstance(x, list) else text_cleaning(x))
# del df['review']
df.head()
```

```python
from konlpy.tag import Okt

def get_nouns_adj(x):
    tagger = Okt()
    if isinstance(x, list):
        x = ' '.join(x)
    pos = tagger.pos(x)
    pos = [word for word, tag in pos if tag not in ['Josa']]
    return pos

df['review'] = df['review'].apply(get_nouns_adj)
# 조사 제거
```

```python
def get_nouns_adj(x):
    tagger = Okt()
    if isinstance(x, list):
        x = ' '.join(x)
    pos = tagger.pos(x)
    pos = [word for word, tag in pos if tag in ['Noun', 'Adjective']]
    return pos

df['review'] = df['review'].apply(get_nouns_adj)
```

```python
# 불용어 리스트
stopwords = ['의', '가', '이', '은', '들', '는', '좀', '잘', '걍', '로', '곳', '요', '게'
             '과', '도', '를', '으로', '자', '에', '와', '한', '하다', '이나', '항상', '것',
            '나', '번', '그', '수']
  # 실제로 사용할 불용어를 여기에 추가하세요.

# 불용어를 제거하는 함수
def remove_stopwords(text):
    return [word for word in text if word not in stopwords]

# 데이터프레임에 함수 적용
df['review'] = df['review'].apply(remove_stopwords)
df
```

```python
from konlpy.tag import Okt

def get_lemmas(text_list):
    okt = Okt()
    result = []
    for text in text_list:
        lemmas = okt.pos(text, norm=True, stem=True)
        lemmas = [lemma[0] if (lemma[1] not in ['Josa', 'PreEomi', 'Eomi', 'Suffix', 'Punctuation']) else '' for lemma in lemmas]
        result.append(' '.join(lemmas))
    return result

# 입력 텍스트 리스트
text_list = df['review'][0]

# 표제어 추출
lemmatized_texts = get_lemmas(text_list)
for text in lemmatized_texts:
    print(text)
```
