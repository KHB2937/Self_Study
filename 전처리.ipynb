```python
# 불용어 리스트
stopwords = ['의', '가', '이', '은', '들', '는', '좀', '잘', '걍', '로', '곳', '요', '게'
             '과', '도', '를', '으로', '자', '에', '와', '한', '하다', '이나']
  # 실제로 사용할 불용어를 여기에 추가하세요.

# 불용어를 제거하는 함수
def remove_stopwords(text):
    return [word for word in text if word not in stopwords]

# 데이터프레임에 함수 적용
df['review'] = df['review'].apply(remove_stopwords)
```

```python
from konlpy.tag import Okt

def get_lemmas(text_list):
    okt = Okt()
    result = []
    for text in text_list:
        lemmas = okt.pos(text, norm=True, stem=True)
        lemmas = [lemma[0] if (lemma[1] not in ['Josa', 'PreEomi', 'Eomi', 'Suffix', 'Punctuation']) else '' for lemma in lemmas]
        result.append(' '.join(lemmas))
    return result

# 입력 텍스트 리스트
text_list = df['review'][0]

# 표제어 추출
lemmatized_texts = get_lemmas(text_list)
for text in lemmatized_texts:
    print(text)
```

```python
negative_words = ['쓴맛', '맛없어요', '맛없', '맛없고', '맛없음', '맛없다는', '맛없으면', '맛있을까', '맛있을수가', '맛없어서', '맛없었어요', '맛없는건',
                  '맛없어', '비맛', '맛없이', '맛없지도', '맛없다도', '맛없네요', '맛있다던데', '맛없어용', '쇠맛', '발맛', '추맛', '맛없는데', 
                  '맛없다에', '트맛', '비릿한맛', '맛없나', '맛없습니다', '절맛', '쟌맛', '햄맛', '맛없었네요', '맛변']

def replace_positive_words(word_list):
    new_list = []
    for word in word_list:
        if word not in negative_words:
            if '맛' in word:
                new_list.append('맛있다')
            else:
                new_list.append(word)
        else:
            new_list.append(word)
    return new_list

df['review'] = df['review'].apply(replace_positive_words)

# Check if the replacement worked
df.head()
```
